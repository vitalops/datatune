```python
from datatune.api import API
from datatune.entity import Entity
from datatune.workspace import Workspace
```

```python
org_name = 'vitalops_org'
```

```python
api = API(base_url='http://localhost:8000', api_key=api_key)
entity = Entity(id=org_name,api=api)
```

```python
ws =  Workspace(id='Awesome Workspace',entity=entity)
print(ws.id)
```

```python
credentials_details = {
    "hf_access_token": "hf_abcdef123456",  # Example credential details
}
```

```python
ws.add_credentials(
    name="HF Credential",
    credential_type="hf",
    credentials=credentials_details,
)
```

```python
dataset1 = ws.add_dataset(
    path="path_to_your_dataset.csv",
    name="My Dataset 1"
)
```

```python
print(dataset1.id)
```

```python
view1 =  ws.create_view(view_name='view1')
```

```python
print(view1.id)
```

```python
view1 = view1.extend(data=dataset1, start=0, stop=100)
```

```python
print(view1.id)
```

```python
view1 = view1.add_extra_column(
    column_name="Example Column 1",
    column_type="str",
    labels=["Label1", "Label2"],
    default_value="Default Value"
)
```

```python
print(view1.id)
```
